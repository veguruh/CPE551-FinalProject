```python
Player_Stats["Player"] = Player_Stats["Player"].str.replace("*", "", regex=False)
```

```python
def single_row(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
players_stats = Player_Stats.groupby(["Player", "Season"]).apply(single_row)
```

```python
players_stats.head(20)
```

```python
players_stats.index = players_stats.index.droplevel()
```

```python
players_stats.index = players_stats.index.droplevel()
```

```python
merged = players_stats.merge(mvps, how="outer", on=["Player", "Season"])
```

```python
Nicknames ={}

with open("Nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        Nicknames[abbrev] = name
```

```python
merged["Team"]=merged["Tm_x"].map(Nicknames)
```

```python
merged.head(10)
```

```python
stats = merged.merge(Teams_Stats, how="outer", on=["Team", "Season"])
```
